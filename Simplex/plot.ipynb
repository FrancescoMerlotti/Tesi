```python
import numpy as np
import matplotlib.pyplot as plt

p0_u = np.loadtxt("output_simplex_u.dat", usecols=0)
p1_u = np.loadtxt("output_simplex_u.dat", usecols=1)
p2_u = np.loadtxt("output_simplex_u.dat", usecols=2)

p0_n = np.loadtxt("output_simplex_n.dat", usecols=0)
p1_n = np.loadtxt("output_simplex_n.dat", usecols=1)
p2_n = np.loadtxt("output_simplex_n.dat", usecols=2)

fig = plt.figure(figsize=[15, 12])

ax1 = fig.add_subplot(1, 2, 1, projection='3d')
ax1.set_title("Simplex uniform distribution", fontsize=18)
ax1.set_xlabel("$\psi_0^2$")
ax1.set_ylabel("$\psi_1^2$")
ax1.set_zlabel("$\psi_2^2$")
ax1.view_init(45, 45)
ax1.scatter3D(p0_u**2, p1_u**2, p2_u**2, color="cornflowerblue")

print("Uniformity check (uniform distribution):")
print("<psi_0^2> =", np.average(p0_u**2))
print("<psi_1^2> =", np.average(p1_u**2))
print("<psi_2^2> =", np.average(p2_u**2))

ax2 = fig.add_subplot(1, 2, 2, projection='3d')
ax2.set_title("Simplex non-uniform distribution", fontsize=18)
ax2.set_xlabel("$\psi_0^2$")
ax2.set_ylabel("$\psi_1^2$")
ax2.set_zlabel("$\psi_2^2$")
ax2.view_init(45, 45)
ax2.scatter3D(p0_n**2, p1_n**2, p2_n**2, color="sandybrown")

print("Uniformity check (non-uniform distribution):")
print("<psi_0^2> =", np.average(p0_n**2))
print("<psi_1^2> =", np.average(p1_n**2))
print("<psi_2^2> =", np.average(p2_n**2))

plt.show()
```
