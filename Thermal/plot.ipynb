```python
import numpy as np
import matplotlib.pyplot as plt

for i in range(10):
	N = i+1
	M = (N+1) / 2

	n = np.loadtxt("./graph_thermal/output_" + str(N) + ".dat", usecols=0)
	h = np.loadtxt("./graph_thermal/output_" + str(N) + ".dat", usecols=1)

	fig = plt.figure(figsize=[8, 6])
	plt.title("Quantum Fisher Information, N = "+str(N))
	plt.xlabel("$N_{tot}$", fontsize=14)
	plt.ylabel("$H$", fontsize=14)
	plt.scatter(n, h, color="cornflowerblue")
	plt.scatter(M, M**2, color="indianred")
	plt.legend(["Thermal", "Vacuum"], loc=0)
	plt.grid()
	plt.savefig("./images_thermal/output_"+str(N)+".jpg", dpi=160)
```

```python
import numpy as np
import matplotlib.pyplot as plt

N = 5

n = np.loadtxt("./graph_thermal/output_" + str(N) + ".dat", usecols=0)
h = np.loadtxt("./graph_thermal/output_" + str(N) + ".dat", usecols=1)
m = np.loadtxt("./graph_thermal/output_" + str(N) + ".dat", usecols=2)

fig = plt.figure(figsize=[8, 6])
plt.title("Quantum Fisher Information, N = "+str(N))
plt.xlabel("$N_{tot}$", fontsize=14)
plt.ylabel("$H$", fontsize=14)
plt.scatter(n, h, color="cornflowerblue")
plt.scatter(n, m, color="indianred")
plt.grid()
plt.plot()
```
